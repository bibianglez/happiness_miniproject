```python
import pycountry
from difflib import get_close_matches

def clean_country_name(country):
    country = country.strip()

    try:
        return pycountry.countries.lookup(country).name
    except LookupError:
        all_country_names = [c.name for c in pycountry.countries]
        match = get_close_matches(country, all_country_names, n=1, cutoff=0.75)

        if match:
            return match[0]
        else:
            return country

population_df["country_clean"] = population_df["country"].apply(clean_country_name)
```

```python
#Checkcing the countries that didn't find a match using the pycountry library:

bad_countries = []

for country in population_df["country_clean"].unique():
    try:
        pycountry.countries.lookup(country)
    except:
        bad_countries.append(country)

bad_countries
```

```python
population_country_fixes = {
    "Republic of Korea": "Korea, Republic of",
    "Russian Federation": "Russia",
    "Cyprus": "North Cyprus",
    "Kosovo (under UNSC res. 1244)": "Kosovo",
    "China, Hong Kong SAR": "Hong Kong",
    "Türkiye": "Turkey",
    "North Macedonia": "Macedonia",
    "Eswatini": "Swaziland",
    "State of Palestine": "Palestinian Territories",
    "Democratic Republic of the Congo": "Congo (Kinshasa)",
    "Congo": "Congo (Brazzaville)",
    "Côte d'Ivoire": "Ivory Coast"
}
```

```python
def fix_population_country(country):
    if country in population_country_fixes:
        return population_country_fixes[country]
    else:
        return country

population_df["country_clean"] = population_df["country_clean"].apply(fix_population_country)
```

```python
population_df[population_df["country_clean"].isin(population_country_fixes.values())]["country_clean"].unique()
```

```python
population_country_fixes = {
    "North Cyprus": "Cyprus",
    "Somalia": "Somaliland region",
    "Congo, The Democratic Republic of the": "Congo (Kinshasa)",
    "Macedonia": "North Macedonia"}
population_df["country_clean"] = population_df["country_clean"].replace(population_country_fixes)
```
