## 🔄 Pipeline overview

### 1. Load the data

Each year is loaded into its own DataFrame (`year_15` through `year_19`) using `pd.read_csv()`.

### 2. Standardize column names

A `clean_column_names()` function strips whitespace, replaces spaces and dots with underscores, and lowercases all column names. A `rename_columns()` function then applies a keyword-based mapping to unify column names across years (since the raw CSVs use different naming conventions each year):

| Normalized name | Examples of raw names matched |
|-----------------|-------------------------------|
| `happiness_score` | `Score`, `Happiness Score` |
| `happiness_rank` | `Rank`, `Happiness Rank` |
| `gdp_percapita` | `Economy (GDP per Capita)`, `GDP per capita` |
| `life_expectancy` | `Health (Life Expectancy)` |
| `family` | `Social support`, `Family` |
| `government_corruption` | `Trust (Government Corruption)` |
| `freedom` | `Freedom to make life choices` |

### 3. Data quality checks

- **Duplicates**: checked across all five DataFrames — none found.
- **Null values**: only one null detected, in the `government_corruption` column for year 2018.

### 4. Clean country names

The `pycountry` library is used to standardize country names to their official ISO forms. A `clean_country_name()` function applies this lookup and creates a new `country_clean` column alongside the original. Countries that don't resolve automatically are fixed manually with a `country_name_fixes` dictionary.

### 5. Reconstruct the `region` column

The `region` column only exists in the 2015 and 2016 datasets. For 2017–2019:

1. A `country_region_dict` is built from the 2015 and 2016 data and mapped onto later years.
2. Countries still missing a region after the map are assigned one manually via `manual_region_fixes`.
3. A final null check confirms all countries have a region before proceeding.

### 6. Add a `year` column

Each DataFrame receives a `year` column with its corresponding integer value (2015–2019).

### 7. Select columns and concatenate

Only a consistent subset of columns is kept across all years:

```python
columns_to_keep = [
    "happiness_rank", "year", "country_clean", "region",
    "happiness_score", "gdp_percapita", "life_expectancy",
    "family", "freedom", "generosity", "government_corruption"
]
```

All five DataFrames are concatenated with `pd.concat()` and `ignore_index=True`. The `country_clean` column is renamed to `country` in the final output.
