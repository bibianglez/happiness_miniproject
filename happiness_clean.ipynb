```python
def clean_column_names(df):
    df.columns = df.columns.str.strip().str.replace(" ", "_").str.replace(".", "_").str.lower()
    return df

clean_column_names(year_15)
clean_column_names(year_16)
clean_column_names(year_17)
clean_column_names(year_18)
clean_column_names(year_19)

column_mapping = {"gdp": "gdp_percapita","health": "life_expectancy","score": "happiness_score","rank": "happiness_rank","corruption": "government_corruption","country": "country","social": "family","freedom": "freedom"}
def rename_columns(df, mapping):
    new_columns = []
    for col in df.columns:
        new_name = col
        for key, value in mapping.items():
            if key in col:
                new_name = value
                break
        new_columns.append(new_name)
    df.columns = new_columns
    return df

year_15=rename_columns(year_15, column_mapping)
year_16=rename_columns(year_16, column_mapping)
year_17=rename_columns(year_17, column_mapping)
year_18=rename_columns(year_18, column_mapping)
year_19=rename_columns(year_19, column_mapping)

display(year_15.columns)
display(year_16.columns)
display(year_17.columns)
display(year_18.columns)
display(year_19.columns)
```

```python
import pycountry
```

```python
#Clean country columns in all data frames using the library 'pycountry':
def clean_country_name(country):
    country = country.strip()
    try:
        return pycountry.countries.lookup(country).name
    except LookupError:
        return country

for df in [year_15, year_16, year_17, year_18, year_19]: df["country_clean"] = df["country"].apply(clean_country_name)
```

```python
# Checking if the region column is replicable for the years that don't include it and if there are duplicates or null values:
display(year_15['region'].value_counts())
display(year_16.region.value_counts())
display(len(year_15.country))
display(len(year_16.country))
display(len(year_17.country))
display(len(year_18.country))
display(len(year_19.country))

# Create a dictionary containing all the countries in the df for year 15 and 16 and drop duplicates:
country_region_map = (pd.concat([year_15[["country_clean", "region"]],year_16[["country_clean", "region"]]]).drop_duplicates().dropna())
country_region_map.groupby("country_clean")["region"].nunique().sort_values(ascending=False)
country_region_dict = country_region_map.drop_duplicates("country_clean").set_index("country_clean")["region"]
```

```python
# Creating the region column for the dataframes year 2016, 2017, 2018:
year_17["region"] = year_17["country_clean"].map(country_region_dict)
year_18["region"] = year_18["country_clean"].map(country_region_dict)
year_19["region"] = year_19["country_clean"].map(country_region_dict)
```

```python
#Checking if after the creation there are any null values in the region column:
display(year_17[year_17["region"].isna()]["country_clean"].unique())
display(year_18[year_18["region"].isna()]["country_clean"].unique())
display(year_19[year_19["region"].isna()]["country_clean"].unique())
```

```python
#Fix manually the missing values which appear to be non corrected countries using the library:
country_name_fixes = {"Taiwan Province of China": "Taiwan, Province of China","Hong Kong S.A.R., China": "Hong Kong","Trinidad & Tobago": "Trinidad and Tobago","Northern Cyprus": "Cyprus","North Macedonia": "North Macedonia","Gambia": "Gambia"}
for df in [year_17, year_18, year_19]: df["country_clean"] = df["country_clean"].replace(country_name_fixes)
```

```python
# Check once the names of countries are corrected if they are still null and if so is a problem with these countries not being in the original dictionary so we will have to add them manually for regions:
display(year_17[year_17["region"].isna()]["country_clean"].unique())
display(year_18[year_18["region"].isna()]["country_clean"].unique())
display(year_19[year_19["region"].isna()]["country_clean"].unique())

manual_region_fixes = {"Taiwan, Province of China": "Eastern Asia","Hong Kong": "Eastern Asia","Trinidad and Tobago": "Latin America and Caribbean","Cyprus": "Middle East and Northern Africa","North Macedonia": "Central and Eastern Europe","Gambia": "Sub-Saharan Africa"}
year_17["region"] = year_17["region"].fillna(year_17["country_clean"].map(manual_region_fixes))
year_18["region"] = year_18["region"].fillna(year_18["country_clean"].map(manual_region_fixes))
year_19["region"] = year_19["region"].fillna(year_19["country_clean"].map(manual_region_fixes))
```

```python
# Checking one final time for nulls:
display(year_17[year_17["region"].isna()]["country_clean"].unique())
display(year_18[year_18["region"].isna()]["country_clean"].unique())
display(year_19[year_19["region"].isna()]["country_clean"].unique())
```

```python
# Create year column in each data framing before we create the final data frame with all the info:
year_15["year"] = 2015
year_16["year"] = 2016
year_17["year"] = 2017
year_18["year"] = 2018
year_19["year"] = 2019
```

```python
# Choosing the columns we want in our finally clean dataframe:
columns_to_keep = ["happiness_rank","year","country_clean","region","happiness_score","gdp_percapita","life_expectancy","family","freedom","generosity","government_corruption"]
final_15 = year_15[columns_to_keep]
final_16 = year_16[columns_to_keep]
final_17 = year_17[columns_to_keep]
final_18 = year_18[columns_to_keep]
final_19 = year_19[columns_to_keep]

hapy_country_df = pd.concat([final_15, final_16, final_17, final_18, final_19],ignore_index=True)
hapy_country_df = hapy_country_df.rename(columns={"country_clean": "country"})
```
