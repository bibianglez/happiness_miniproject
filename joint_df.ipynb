```python
import pandas as pd

population_df = pd.read_csv("population_df.csv")
hapy_country_df=pd.read_csv("hapy_country_df.csv")
```

```python
display(hapy_country_df.head)
display(population_df.head)
```

```python
happy_population_df = hapy_country_df.merge(population_df[["country_clean", "year", "total_pop", "pop_density", "age_pop"]]
    ,left_on=["country", "year"],
    right_on=["country_clean", "year"],
    how="left")
```

```python
happy_population_df = happy_population_df.drop(columns=["country_clean"])
```
