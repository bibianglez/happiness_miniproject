```python
import pandas as pd
import numpy as np

df = pd.read_csv("HAPPY_POPULATION.csv")
```

```python
# TOP 10 COUNTRIES BY HAPPINESS SCORE:

country_happiness = df.groupby('country')['happiness_score'].mean().sort_values(ascending=False)
display(country_happiness.head(10))
country_happiness.head(10).plot(kind='bar', figsize=(12, 6),title='Top 10 Countries by Mean Happiness Score (2015–2019)')
```

```python
# TOP COUNTRIES BY HAPPINESS SCORE:

top10_regions = df.groupby('region')['happiness_score'].mean().sort_values(ascending=False).head(10)

top10_regions.plot(kind='barh', figsize=(12, 6),title='Top 10 Regions by Mean Happiness Score (2015–2019)')
```

```python
# DESCRIPTIVE METHODS ON HAPPINESS VARIABLES:

def descriptive_stats_simple(data, variables):
    rows = []
    for v in variables:
        s = data[v].dropna()
        mode_val = s.mode()
        mode_str = ", ".join(f"{m:.4f}" for m in mode_val[:3])
        rows.append({
            "Variable": v,
            "Mean": round(s.mean(), 4),
            "Median": round(s.median(), 4),
            "Mode": mode_str,
            "Min": round(s.min(), 4),
            "Max": round(s.max(), 4), })
    return pd.DataFrame(rows).set_index("Variable")

HAPPINESS_VARS = ["happiness_score","gdp_percapita","life_expectancy","family","freedom","generosity","government_corruption"]

stats_happiness = descriptive_stats_simple(df, HAPPINESS_VARS)
print(stats_happiness.to_string())
```

```python
POPULATION_VARS = ["total_pop","pop_density","age_pop"]
stats_pop = descriptive_stats_simple(df, POPULATION_VARS)
print(stats_pop.to_string())
```

```python
# HAPPINESS SCORE BY REGION BY YEAR

pivot_region_year = df.pivot_table(index='year',columns='region',values='happiness_score',aggfunc='mean')

pivot_region_year.plot(figsize=(12, 6),marker='o',title='Happiness score by Year for Regions')
```

```python
#Full correlation matrix — all happiness variables
corr_happiness = df[HAPPINESS_VARS].corr().round(3)
display(corr_happiness)
```

```python
corr_pop_vs_happiness = (df[["happiness_score"] + POPULATION_VARS].corr().round(3))
display(corr_pop_vs_happiness)
```
